```python
import matplotlib.pyplot as plt #so we can plot
import numpy as np #so we can calculate

def gradient_descent (f, learning_rate, ini_point):

    x_coords=[ini_point]
    y_coords=[f(ini_point)]

    def deriv(x): #estimate derivative of function
        return (f(x+0.00001) -f(x))/0.00001

    i = 1 #index of 1
    while abs(deriv(x_coords[i-1]))>0.00001: #loop runs until there is a value 0.00001 away from the local minima

        xi=x_coords[i-1]-learning_rate*deriv(x_coords[i-1])

        x_coords.append(xi)
        y_coords.append(f(x_coords[i])) #everytime you run through loop, append point to array before recalculating
        i = i + 1

    plot_range=np.linspace(min(x_coords)-0.1, max(x_coords)+0.1,10000)

    fxn_range=[f(i) for i in plot_range]
    plt.plot(plot_range, fxn_range)
    plt.plot(x_coords,y_coords)

    return round (x_coords[-1],3),round(y_coords[-1],3)
```

```python
def func_a(x):
    return x**2

gradient_descent (func_a, 0.9, 1)
```

```python
def func_b(x):
    return (x**3)*-3 + 10*x**2 - 5*x -3

gradient_descent (func_b, 0.104, -0.01)
```
